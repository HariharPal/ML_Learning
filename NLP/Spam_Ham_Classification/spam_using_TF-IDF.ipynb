## Fetching the data

```python
import pandas as pd
messages = pd.read_csv("spam.csv", encoding="utf-8", encoding_errors="ignore")
```

```python
messages
```

```python
type(messages)
```

```python
messages = messages.rename(columns={"v1" : "Type", "v2": "Message"})
```

```python
messages
```

```python
messages.drop(columns={"Unnamed: 2", "Unnamed: 3", "Unnamed: 4"}, inplace=True)
```

```python
messages
```

## Data Cleaning

```python
import re
import nltk  
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub("[^a-zA-Z]", " ", messages["Message"][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words("english")]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Output feature

```python
y = pd.get_dummies(messages["Type"])
y
```

```python
y = y.iloc[:,0].values
```

```python
y
```

```python
y = y.astype(int)
```

## Train test split


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus,y, test_size=0.23)
```

```python
y_train
```

## Creating TF-IDF & NGrams

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=1500, ngram_range=(1,2))
```

```python
X_train = tfidf.fit_transform(X_train).toarray()
X_test = tfidf.transform(X_test).toarray()
```

## Appling Training & Testing

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spamDetectModel = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred = spamDetectModel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
