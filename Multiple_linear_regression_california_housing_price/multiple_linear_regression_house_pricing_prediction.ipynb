```python
#Lets prepare dataframe 
dataset = pd.DataFrame(california.data, columns=california.feature_names)
dataset.head()#all independent features which define or describe the pricing of houses is present in dataset now
```

```python
#Lets add dependent feature
dataset['Price'] = california.target
```

```python
dataset.head()#We see price will be added in our dataset
```

```python
dataset.isnull().sum()#No missing value hence no need for any mean median or mode imputation
```

```python
import seaborn as sns
sns.heatmap(dataset.corr(  ), annot= True)
```

```python
dataset.head()
```

```python
##Dividing data into independent and dependent features
X = dataset.iloc[:, : -1] # We say take all rows or value and get all feature except the last one so we're subtracting wiht -1; Independent features
y = dataset.iloc[: , -1]#Dependent features
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=10 )
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#Standerdization of the dataset 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train  = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
##Model training
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train, y_train)
```

```python
#Slope or coeff
regression.coef_
```

```python
regression.intercept_
```

```python
##Prediction of test data
y_pred = regression.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
rmse = np.sqrt(mse)
print("MSE", mse)
print("MAE", mae)
print("RMSE", rmse)
```

```python
#Finding r2 square
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
score
```

```python
#Adjusted r2 
1-(1-score) * (len(y_test) -1) / (len(y_test) - X_test.shape[1]-1)
```

```python
#Assumptions
plt.scatter(y_test, y_pred)
plt.xlabel("True data")
plt.ylabel("Predicted data")
```

# Pickling  
Python pickle module is used to serializing and deserializing a python object structure. It is way to convert python object (list , dict , etc) into charactor stream.The idea is that this stream contains all the information neccessory to reconstruct the object in another python script. 

```python
import pickle 
pickle.dump(regression, open('regressor.pkl', 'wb'))#Utf 8 encoded
```

```python
model = pickle.load(open('regressor.pkl', 'rb'))#importing the pickle(aachar);
```

```python
model.predict(X_test)
```
