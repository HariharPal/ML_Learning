```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = 6 * np.random.rand(100, 1) -3
y= 0.5*X**2 + 1.5*X + 2 + np.random.randn(100, 1)
#Quardratic equation use y = .5X^2 + 1.5X + 2 + outliers
```

```python
plt.scatter(X,y, color ='g')
plt.xlabel("X Dataset")
plt.ylabel("y Dataset")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X, y,test_size= 0.25, random_state=42)
```

```python
#Lets implement Simple linear regression 
from sklearn.linear_model import LinearRegression
regression_1 = LinearRegression()
```

```python
regression_1.fit(X_train , y_train)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, regression_1.predict(X_test))
```

```python
score
```

```python
#lets visulaize the model
plt.plot(X_train, regression_1.predict(X_train), color = "r" )
plt.scatter(X_train , y_train)
plt.xlabel("X dataset")
plt.ylabel("y dataset")
```

```python
#Lets apply polynomial transformation
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(include_bias= True, degree= 2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
from sklearn.metrics import r2_score
regression = LinearRegression()
regression.fit(X_train_poly, y_train)
y_pred = regression.predict(X_test_poly)
score = r2_score(y_test, y_pred)
score
```

```python
print(regression.coef_)
print(regression.intercept_)
```

```python
plt.scatter(X_train, regression.predict(X_train_poly))
plt.scatter(X_train, y_train)
```

```python
#Polynomail As degree 3
poly = PolynomialFeatures(include_bias= True, degree= 3)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
from sklearn.metrics import r2_score
regression = LinearRegression()
regression.fit(X_train_poly, y_train)
y_pred = regression.predict(X_test_poly)
score = r2_score(y_test, y_pred)
score
```

```python
print(regression.coef_)
print(regression.intercept_)
```

```python
plt.plot(X_train, regression.predict(X_train_poly), "r.",label = "Predictions")
plt.plot(X_train, y_train, 'b.')
plt.legend()
plt.show()
```

```python
#Prediction of new Dataset
X_new = np.linspace(-3, 3,200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
```

```python
y_new = regression.predict(X_new_poly)
plt.plot(X_new, y_new, 'r-', linewidth= 2, label= "New Predictions")
plt.plot(X_train, y_train, 'b.' , label= "Training points")
plt.plot(X_test , y_test, 'g.' , label = "Testing Points")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

Pipelining Concept

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    X_new = np.linspace(-3, 3, 200).reshape(200, 1)
    X_new_poly = poly.transform(X_new)
    poly_features = PolynomialFeatures(degree= degree, include_bias=True)
    lin_reg = LinearRegression()
    poly_regression = Pipeline([("poly_features" , poly_features), ("lin_reg" , lin_reg)])
    poly_regression.fit(X_train, y_train)
    y_pred_new = poly_regression.predict(X_new)
    #Plotting prediction line
    plt.plot(X_new, y_pred_new , 'r', label= "Degree "+str(degree), linewidth = 3)
    plt.plot(X_train , y_train , 'b.', linewidth = 3)
    plt.plot(X_test, y_test, 'g.' , linewidth = 3)
    plt.legend (loc = "upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-4, 4,0, 10])
    plt.show()
```

```python
poly_regression(2)
```
